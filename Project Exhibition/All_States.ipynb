```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
%matplotlib inline
```

## Tamil Nadu
Block: Maduranthakam

```python
tamil_nadu = pd.read_csv("./All States CSV/Tamil Nadu.csv")
tamil_nadu.head()
```

```python
Cat_of_Schools= sns.countplot(data=tamil_nadu, x="school_subcategory")
Cat_of_Schools.set_xticklabels(Cat_of_Schools.get_xticklabels(), rotation=40, ha="right")
```

```python
Types_of_School = sns.barplot(data=tamil_nadu, x="school_name", y="no_of_student")
Types_of_School.set_xticklabels(Types_of_School.get_xticklabels(), rotation=40, ha="right", fontsize=6)
```

```python
tamil_nadu["school_category"].value_counts()
```

## Kerala
Block Name: Kanhangad

```python
kerala = pd.read_csv("./All States CSV/Kerala.csv")
kerala.head()
```

```python
Cat_of_Schools= sns.countplot(data=kerala, x="school_subcategory")
Cat_of_Schools.set_xticklabels(Cat_of_Schools.get_xticklabels(), rotation=40, ha="right")
```

```python
Types_of_School = sns.barplot(data=kerala, x="school_name", y="no_of_student")
plt.figure(figsize=(72, 48))
Types_of_School.set_xticklabels(Types_of_School.get_xticklabels(), rotation=40, ha="right", fontsize=6)
```

```python
kerala["school_category"].value_counts()
```
