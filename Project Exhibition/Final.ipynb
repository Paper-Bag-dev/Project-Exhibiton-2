```python
Schools_with_tap = tb.read_pdf("DataTable.pdf", pages="all")
```

```python
Schools_with_tap
```

```python
df = pd.DataFrame(Schools_with_tap[0])
df.head(10)
```

```python
df.rename(columns={"Schools\rwith tap\rwater supply (%)\r" : "Schools with tap water supply in %"}, inplace=True)
df
```

```python
print(df.columns)
```

```python
df.rename(columns={'State/ UT\r?' : 'State/UT', 
                   'Total\rschools\r?' : 'Total Schools',
                   'Schools\rwith tap\rwater supply\r?' : 'Schools with tap water supply',
                   'Schools\rwith tap\rwater supply (%)\r?' : 'Schools with tap water supply in %'}, inplace=True)

# Plot the data
plt.plot(df['State/UT'], df['Schools with tap water supply in %'])
plt.title('Proportion of Schools with Tap Water Supply')
plt.xlabel('States')
plt.ylabel('Percentage of Schools with Tap Water Supply')
plt.xticks(rotation=90)
plt.show()
```

```python
df = df.iloc[:36, :4]

df.rename(columns={'State/ UT\r?' : 'State/UT', 'Total\rschools\r?' : 'Total Schools'}, inplace=True)

plt.bar(df['State/UT'], df['Total Schools'])
plt.title('Total Number of Schools in Each State')
plt.xlabel('States/UTs')
plt.ylabel('Total Number of Schools')
plt.xticks(rotation=90)
plt.show()
```
